```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#import data with formatting the dataframe
mushroom = pd.read_fwf('agaricus-lepiota.data', delimiter =",",header=None)
mushroom
```

```python
#capture columns.
#rename columns.
#replace the value with numbers.
new= mushroom[[0,5,8]]
new.columns = ['classes','odor','size']
new= new.replace({'classes': {'e': 0, 'p': 1},
             'odor':{'n':0,'a':1,'l':2,'c':3,'y':4,'f':5,'m':6,'p':7,'s':8},
             'size':{'b':0,'n':1}})
new.astype(int)
```

```python
classplot = sns.countplot(x="classes", data=new)
classplot.set(xlabel='Edible = 0, Poisonous = 1')
```

```python
odorplot = sns.countplot(x="odor", data=new)
odorplot.set(xlabel='None = 0, Almond = 1, Anise = 2, Creosote = 3, Fishy = 4, Foul = 5, Musty = 6, Pungent = 7, Spicy = 8')
```

```python
sizeplot = sns.countplot(x="size", data=new)
sizeplot.set(xlabel='Broad = 0, Narrow = 1')
```

```python
sns.scatterplot(data=new, x="odor", y="size", hue="classes")
```

From the above graphs, the amount of Edible and Poisonous species are almost half half. Most of mushrooms have no odor, followed by foul smell. And most are narrow in size.
From last graph, we can find that mushrooms with Almond and Anise smells are all ediable not matter narrow or broad in size. The None smell mushroom is also edible but only narrow size. all other smells no matter narrow or broad are poisonous. 
By using scikit-learn method, other columns do give more insights to predict if the mushroom is edible or not. The narrow size mushrooms are more likely to be poisonous compare to broad. And the smell is more strong, the mushroom is more likely to be posionous.
