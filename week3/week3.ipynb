import all necessary modules

```python
import pandas as pd
from numpy import random
import matplotlib.pyplot as plot
from matplotlib.ticker import MaxNLocator
```

generate daily mileages for next 18 days, (assuming total spend 20 days.)

```python
random.seed(1000)
randomdaily = [random.randint(low=45,high=75) for i in range(18)]
print(randomdaily)
```

combine first two days data

```python
daily = [55,65]
daily += randomdaily
print (daily)
```

calcuate odometer for all 20 days

```python
odometer = []
total = 0
for i in daily: 
    total += i
    odometer.append(total)
print(odometer)
```

create a list for date

```python
days = list(range(1,len(odometer)+1))
print(days)
```

zip all three data into dataset

```python
datacombination = list(zip(days,daily,odometer))
print(datacombination)
```

```python
df = pd.DataFrame(data = datacombination, columns=['Days', 'Miles', 'Total Miles'])
df
```

daily mileage bar chart

```python
df['Miles'].plot.bar()
plot.suptitle('Daily Mileage')
plot.ylabel('Miles')
plot.xlabel('Day')
```

daily odometer graph

```python
df['Total Miles'].plot()
plot.suptitle('Odomter')
plot.ylabel('Daily Odometer')
plot.xlabel('Day')
plot.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plot.grid()
```
