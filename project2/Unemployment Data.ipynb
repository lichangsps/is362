```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
unemployment = pd.read_csv('Unemployment Data.csv', index_col =0)
unemployment.head()
```

```python
unemployment.shape
```

the dataframe is too wide, I decided to go long instead of wide. 
I use transpose function

```python
T_unemployment = unemployment.T
T_unemployment.head()
```

```python
T_unemployment.shape
```

The dataframe has been tranposed, but the first column is NaN, 
I decided to check the NaN across the dataframe and clean their as necessary.

```python
T_unemployment.isna().sum()
```

only one column is NaN, with none values. 
also notice that all column names (year) have been changed to float type.

```python
T_unemployment=T_unemployment.dropna(axis=1, how='all')
T_unemployment.columns = T_unemployment.columns.astype(int)
T_unemployment
```

I am interested in the unempolyment information during the 2008 economic crisis. I would recall 2007 to 2011 data to compare.

```python
Unemploy = T_unemployment.iloc[:, 17:22]
Unemploy
```

Notice there are some NaN value in this frame, I would exclude all rows with NaN.

```python
Unemploy=Unemploy.dropna()
Unemploy.isnull().sum()
```

```python
Unemploy
```

This dataframe calcuates mean, std, min, 25%, 50%, 75% and max value for every years.  

```python
Unemploy.describe()
```

Compare the mean values, we can see the unemployment rate reach to the highest in 2010.

```python
Unemploy.describe().iloc[1].plot(kind = 'bar')
```

Top 5 countries/regions with the highest unemployment rate. compare by average of 5 years data.

```python
mean = Unemploy.agg("mean", axis="columns")
mean.sort_values(ascending=False).head()
```
