```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
pd.options.display.float_format = "{:,.2f}".format
```

the data is composed by two parts, first part is two year school tuition and second is four year school tuition.
I will sperate them into two tables.

```python
df = pd.read_csv('Tuition fee by state.csv', index_col =0, header=2)
df.index.names = ['State']
df
```

Check NaN data information, notice first part has 3 nulls and none for second part

```python
df.isnull().sum()
```

split data and rename column names.

```python
two_year_orgin= df.iloc[:,:14]
two_year_orgin.columns = ['2004-05','2005-06','2006-07','2007-08','2008-09','2009-10','2010-11','2011-12','2012-13',
'2013-14','2014-15','2015-16','2016-17','2017-18']
four_year_orgin= df.iloc[:,17:31]
four_year_orgin.columns = ['2004-05','2005-06','2006-07','2007-08','2008-09','2009-10','2010-11','2011-12','2012-13',
'2013-14','2014-15','2015-16','2016-17','2017-18']
```

all data are str type because they contain '$' and ',', cannot directly process the data. 
converting to float type by removing '$'and ','

```python
two_year_tuition = two_year_orgin
a1 = list(two_year_tuition.columns)
two_year_tuition[a1]= two_year_tuition[a1].astype(str)  
for i in a1:
    two_year_tuition[i] = two_year_tuition[i].str.replace(',', '').str.replace('$', '').astype(float)
    
```

```python
four_year_tuition = four_year_orgin.dropna()
b1 = list(four_year_tuition.columns)
four_year_tuition[b1]= four_year_tuition[b1].astype(str)  
for i in b1:
    four_year_tuition[i] = four_year_tuition[i].str.replace(',', '').str.replace('$', '').astype(float)
```

Generate descriptive statistics and the percentage changes for the statistic

```python
two_year_tuition.describe()
```

```python
two_year_tuition.describe().pct_change(axis='columns')
```

```python
four_year_tuition.describe()
```

```python
four_year_tuition.describe().pct_change(axis='columns')
```

calcuate the tuition diff between four year school and two year school.
join and graph the data

```python
mean_diff = four_year_tuition.mean()/two_year_tuition.mean()
max_diff = four_year_tuition.max()/two_year_tuition.max()
min_diff = four_year_tuition.min()/two_year_tuition.min()
```

```python
mean_diff.name='mean_diff'
max_diff.name='max_diff'
min_diff.name='min_diff'
diff = mean_diff.to_frame().join([max_diff,min_diff])
diff
```

From the graph, overall the different between max tuition and mean tuition are relatively stable, fluctuations are not changing too much. 
Different for minimum tuition is unstable

```python
diff.plot()
```
